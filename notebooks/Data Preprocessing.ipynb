```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
```

```python
def process_str(value):
    # split into words
    tokens = [w.lower() for w in word_tokenize(value)]
    # remove punctuation from each word
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    words = [word for word in stripped if word.isalpha()]
    stop_words = set(stopwords.words('english'))
    return [w for w in words if not w in stop_words]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(lis):
    res = []
    for l in lis.split():
        res.append(ps.stem(l))
    return " ".join(res)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity[1]
```

```python
def recomend(movie):
    movie_idx = df[df['title'] == movie].index[0]
    distences = sorted(list(enumerate(similarity[movie_idx])), reverse=True, key=lambda x: x[1])[1:6]
    for i in distences:
        print(df.iloc[i[0]].title)
```

```python
recomend('John Carter')
```
